```python
# Line plot
import matplotlib.pyplot as py
x=[1,2,3,4,5]
y=[6,7,8,9,10]
py.plot(x,y)
```

```python
# Line plot by using range
import matplotlib.pyplot as p
x=list(range(1,10,2)) #[expression loop condition]
y=list(range(20,30,2))
p.plot(x,y,'s')
p.show()
```

```python
# Lineplot by using list comphrension for 1-10 integers
import math as m
import matplotlib.pyplot as g
x=[i for i in range(1,11)]
x
y=[m.sqrt(i) for i in range(1,11)]
y
g.plot(x,y,'y')
g.grid()
```

```python
import matplotlib.pyplot as d
x=[2,4]
y=[2,0]
d.plot(x,y)
d.xlabel('X-Data')
d.ylabel('y-Data')
d.title('Line plot')
d.grid()
d.show()
```

```python
# Bar graph
import matplotlib.pyplot as f
x=list(range(1,10,3))
print(x)
y=[3,5,8]
f.bar(x,y)
f.show()
```

```python
# Scatter 
# Histogram
```

```python
help(f.hist)
```

```python
x=[2,4,6,8,10]
y=[1,3,5,7,9]
f.scatter(x,y,cmap='hot')# or f.scatter(x,y)
f.title('Scatter')
f.xlabel('X-Axis')
f.ylabel('Y-Axis')
f.show()
```

```python
f.hist(y)
f.title('Histogram')
f.xlabel('X-Axis')
f.ylabel('Y-Axis')
f.show()
```

```python
# Draw a plot time and distance by using list comphrension, distance=2*time
x=[i for i in range(20,30)]
y=[i*2 for i in range(20,30)]
f.plot(x,y)
f.xlabel('X-Time')
f.ylabel('Y-Distance')
f.show()
```
