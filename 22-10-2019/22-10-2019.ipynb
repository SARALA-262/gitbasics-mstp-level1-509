### Matplotlib ticks
* Ticks is values within a coordinates

```python
import matplotlib.pyplot as s
x=[5,10,15,20,25,30,35,40,45,50]
print(x)
y=[9,8,5,6,3,2,1,4,7,10]
print(y)
s.plot(x,y)
s.title('A Graph a with example to ticks')
s.xlabel('Natural numbers')
s.ylabel('Values with ++5 up to 50')
s.show()
```

```python
import matplotlib.pyplot as s
import numpy as np
s.xticks(np.arange(10,20,1))
s.yticks(np.arange(5,51,5))
s.plot(x,y)
print(x)
print(y)
s.title('Matplotlib-Ticks')
s.xlabel('10 natuaral numbers')
s.ylabel('5-50 values with 5+')
s.show()
```

### Representing a sine wave
* A=sine(wt)
* A=Amplitude
* w=frequwncy
* t=time

```python
import numpy as np
import matplotlib.pyplot as s
Time=np.arange(1,10,0.1)
Amplitude=np.sin(Time)
s.grid()
s.plot(Time,Amplitude)
s.title('Sine Wave A=sine(Wt)')
s.xlabel('Time')
s.ylabel('Amplitude')
s.show()
```
