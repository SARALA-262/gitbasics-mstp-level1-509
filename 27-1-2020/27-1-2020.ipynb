```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.set(style="darkgrid")
t=sns.load_dataset('tips')
t.head
```

```python
sns.relplot(x="total_bill",y="day",data=t)
```

```python
sns.relplot(x="total_bill",y="day",data=t,hue="tip")
```

```python
sns.relplot(x="total_bill",y="size",data=t,hue="tip",kind="line")
```

```python
sns.set(style="darkgrid")
m=sns.load_dataset('iris')
m.head()
```

```python
sns.relplot(x="species",y="petal_length",data=m)
```

```python
sns.relplot(x="species",y="petal_length",col="species",data=m)
```

```python
sns.relplot(x="species",y="petal_length",col="species",data=m,hue="species")
```

```python
sns.set(style="white")
i=sns.load_dataset('iris')
i.head()
```

```python
i["species"].value_counts()
```

```python
sns.relplot(x="sepal_width",y="petal_width",data=i)
```

```python
sns.relplot(x="sepal_width",y="petal_width",data=i,hue="species",col="species")
```

```python
#categarical data plot
#catplot
    #SWARM plot
    #SCATTER plot
    #BAR plot
    #Violin plot
```

```python
sns.catplot(x="species",y="petal_length",data=i)
```

```python
sns.catplot(x="species",y="petal_length",hue="sepal_width",data=i)
```

```python
sns.catplot(x="species",y="petal_length",hue="sepal_width",data=i,kind="swarm")
```

```python
sns.catplot(x="species",y="petal_length",data=i,kind="violin")
```

```python
sns.barplot(x="species",y="petal_length",data=i)
```

```python
#box plot
f,axs=plt.subplots(1,2)
sns.boxplot(x="species",y="sepal_length",data=i,ax=axs[0])
```

```python
f,axs=plt.subplots(1,2)
sns.boxplot(x="species",y="sepal_length",data=i,ax=axs[0])
sns.scatterplot(x="species",y="sepal_length",data=i,ax=axs[1])
plt.show()
```
