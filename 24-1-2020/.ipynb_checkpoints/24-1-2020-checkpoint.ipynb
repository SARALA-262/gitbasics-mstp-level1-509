```python
#creatin subplot using numpy with sine
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,'bo')
```

```python
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s)
plt.plot("t")
plt.plot("s")
plt.title("graph t vs s")
```

```python
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,label="graph with sine wave")
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
plt.show()
```

```python
#with non values
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.exp(-t)*np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,label="1st graph with sine wave")
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
t[0]=np.nan
t[1]=np.nan
plt.subplot(2,1,2)
plt.plot(t,s,label="2st graph with sine wave")
plt.xlabel("t")
plt.ylabel("s")
```

```python
x=[3,4,5,6]
y=[1,2,3,4]
z=[6,7,8,9]
dx=np.ones(4)
dy=np.zeros(4)
dz=np.ones(4)
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.bar3d(x,y,z,dx,dy,dz,color='y')
plt.xlabel("x")
plt.ylabel("y")
plt.title("3d project")
plt.show()
```
